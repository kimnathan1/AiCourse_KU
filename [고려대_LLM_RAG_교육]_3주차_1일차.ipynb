```python
import json
import yfinance as yf
from openai import OpenAI

client = OpenAI()

# 1) í™˜ìœ¨ ì¡°íšŒ í•¨ìˆ˜
def get_exchange_rate(currency="KRW"):
    data = yf.Ticker(f"USD{currency}=X").history(period="1d")
    rate = float(data["Close"].iloc[-1])
    return json.dumps({"rate": round(rate, 2), "currency": currency})

# 2) í•¨ìˆ˜ ìŠ¤í‚¤ë§ˆ
TOOLS = [{
    "type": "function",
    "function": {
        "name": "get_exchange_rate",
        "description": "USD ê¸°ì¤€ í™˜ìœ¨ ì¡°íšŒ",
        "parameters": {
            "type": "object",
            "properties": {"currency": {"type": "string", "description": "í†µí™”ì½”ë“œ (KRW, JPY, EUR ë“±)"}},
            "required": ["currency"]
        }
    }
}]

# 3) ìë™ ì‹¤í–‰
def ask(query: str):
    messages = [{"role": "user", "content": query}]

    while True:
        resp = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages,
            tools=TOOLS,
            tool_choice="auto",      # âœ… ëª¨ë¸ì´ í•„ìš”ì‹œ í•¨ìˆ˜ í˜¸ì¶œ ê²°ì •
            temperature=0.2          # âœ… ì‚¬ì‹¤ ê¸°ë°˜ ìš©ë„: ë‚®ê²Œ
        )
        msg = resp.choices[0].message

        # ë„êµ¬ í˜¸ì¶œ ì—†ìœ¼ë©´ ì¢…ë£Œ (ìµœì¢… ìì—°ì–´ ë‹µë³€)
        if not msg.tool_calls:
            return msg.content

        # ì–´ì‹œìŠ¤í„´íŠ¸ ë©”ì‹œì§€(í˜¸ì¶œ ì˜ë„) ì¶”ê°€
        messages.append({
            "role": "assistant",
            "content": msg.content or "",
            "tool_calls": msg.tool_calls
        })

        # ë„êµ¬ ì‹¤í–‰ â†’ ê²°ê³¼ë¥¼ tool ë©”ì‹œì§€ë¡œ ì²¨ë¶€(ì¤‘ìš”: tool_call_id)
        for tc in msg.tool_calls:
            args = json.loads(tc.function.arguments or "{}")
            result = get_exchange_rate(**args)
            messages.append({
                "role": "tool",
                "tool_call_id": tc.id,          # âœ… í•„ìˆ˜
                "name": tc.function.name,       # (ì„ íƒ) ê°€ë…ì„±
                "content": result               # JSON ë¬¸ìì—´
            })

# ì‹¤í–‰ ì˜ˆì‹œ
if __name__ == "__main__":
    print(ask("ì§€ê¸ˆ USD->KRW í™˜ìœ¨ ì•Œë ¤ì¤˜"))
```

```python
# ë©€í‹° Function Calling ì±—ë´‡ â€” ì½”ë“œë§Œ (Colab/ë¡œì»¬ ê³µí†µ)
from openai import OpenAI
import json

client = OpenAI()

# ===== 1) í•¨ìˆ˜ ì •ì˜ =====
def get_weather(city: str):
    return f"{city}ì˜ ë‚ ì”¨ëŠ” ë§‘ê³  22ë„ì…ë‹ˆë‹¤."

def add_schedule(date: str, task: str):
    return f"{date}ì— '{task}' ì¼ì •ì„ ë“±ë¡í–ˆìŠµë‹ˆë‹¤."

def get_study_advice(topic: str):
    return f"'{topic}' ê³µë¶€ëŠ” í•µì‹¬ ê°œë…ì„ ìš”ì•½í•˜ê³  30ë¶„ ë³µìŠµí•˜ëŠ” ê²Œ ì¢‹ì•„ìš”."

# ===== 2) í•¨ìˆ˜ ìŠ¤í‚¤ë§ˆ =====
functions = [
    {
        "name": "get_weather",
        "description": "ë„ì‹œì˜ ë‚ ì”¨ë¥¼ ì¡°íšŒí•©ë‹ˆë‹¤.",
        "parameters": {
            "type": "object",
            "properties": {
                "city": {"type": "string", "description": "ë„ì‹œ ì´ë¦„"}
            },
            "required": ["city"]
        }
    },
    {
        "name": "add_schedule",
        "description": "ìƒˆë¡œìš´ ì¼ì •ì„ ë“±ë¡í•©ë‹ˆë‹¤.",
        "parameters": {
            "type": "object",
            "properties": {
                "date": {"type": "string", "description": "ë‚ ì§œ (ì˜ˆ: 2025-10-21)"},
                "task": {"type": "string", "description": "ì¼ì • ë‚´ìš©"}
            },
            "required": ["date", "task"]
        }
    },
    {
        "name": "get_study_advice",
        "description": "ê³µë¶€ ì£¼ì œì— ë§ëŠ” í•™ìŠµ ì¡°ì–¸ì„ ì œê³µí•©ë‹ˆë‹¤.",
        "parameters": {
            "type": "object",
            "properties": {
                "topic": {"type": "string", "description": "ê³µë¶€ ì£¼ì œ"}
            },
            "required": ["topic"]
        }
    }
]

# ===== 3) ì‹¤í–‰ê¸°: í•¨ìˆ˜ ìë™ ì„ íƒ â†’ ì‹¤í–‰ â†’ ìì—°ì–´ ì‘ë‹µ =====
def route_and_execute(user_input: str, temperature: float = 0.2) -> str:
    # (A) ëª¨ë¸ì´ í•¨ìˆ˜ ìë™ ì„ íƒ
    resp = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": user_input}],
        functions=functions,
        function_call="auto",
        temperature=temperature
    )
    msg = resp.choices[0].message

    # í•¨ìˆ˜ ì„ íƒ ì—†ìœ¼ë©´ ë°”ë¡œ ë‹µë³€ ë°˜í™˜
    if not getattr(msg, "function_call", None):
        return msg.content or ""

    # (B) ì„ íƒëœ í•¨ìˆ˜ ì‹¤í–‰
    func_name = msg.function_call.name
    args = json.loads(msg.function_call.arguments or "{}")

    if func_name == "get_weather":
        result = get_weather(**args)
    elif func_name == "add_schedule":
        result = add_schedule(**args)
    elif func_name == "get_study_advice":
        result = get_study_advice(**args)
    else:
        result = "í•´ë‹¹ ì‘ì—…ì„ ì²˜ë¦¬í•  í•¨ìˆ˜ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."

    # (C) í•¨ìˆ˜ ê²°ê³¼ë¥¼ ëª¨ë¸ì— ì „ë‹¬ â†’ ìì—°ì–´ ìš”ì•½ ì‘ë‹µ
    followup = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": user_input},
            {"role": "assistant", "function_call": msg.function_call},
            {"role": "function", "name": func_name, "content": result}
        ],
        temperature=temperature
    )
    return followup.choices[0].message.content

# ===== 4) í…ŒìŠ¤íŠ¸ =====
if __name__ == "__main__":
    samples = [
        "ë‚´ì¼ ì„œìš¸ ë‚ ì”¨ ì•Œë ¤ì¤˜",
        "2025-10-25ì— íšŒì˜ ì¼ì • ë“±ë¡í•´ì¤˜",
        "SQL ê³µë¶€ ì˜í•˜ëŠ” ë°©ë²• ì•Œë ¤ì¤˜"
    ]
    for s in samples:
        print("\n==============================")
        print("ğŸ‘© ì‚¬ìš©ì:", s)
        print("ğŸ¤– ì‘ë‹µ:", route_and_execute(s))
```

```python
import json, textwrap
from openai import OpenAI

client = OpenAI()

# ===== ìƒíƒœ ê´€ë¦¬ =====
state = {
    "turn": 0,        # ì‚¬ìš©ì ë°œí™” íšŸìˆ˜
    "memory": ""      # 3~5ì¤„ ìš”ì•½ ë©”ëª¨ë¦¬(ëª©í‘œ/ì œì•½/ì„ í˜¸/ì§„ì²™ ìœ„ì£¼)
}

# ===== ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ =====
SYSTEM_PROMPT = """
ë„ˆëŠ” ë”°ëœ»í•˜ê³  ì „ë¬¸ì ì¸ í•œêµ­ì–´ ì§„ë¡œ ìƒë‹´ ì½”ì¹˜ì•¼.
ëŒ€í™” ë‹¨ê³„ëŠ” ë‘ ê°€ì§€ì•¼.

[Phase 1: ìµœì´ˆ 1í„´]
- ì‚¬ìš©ìì˜ ì§„ë¡œ ëª©í‘œì™€ ìƒí™©ì„ íŒŒì•…í•˜ê¸° ìœ„í•œ "ì´ˆê¸° í”„ë¡œí•„ JSON"ë§Œ ë‹µí•´.
- ì¶œë ¥ í˜•ì‹(ì˜¤ì§ JSONë§Œ):
{
  "goal": "ì‚¬ìš©ìì˜ 1ì°¨ ì§„ë¡œ ëª©í‘œ í•œ ë¬¸ì¥ ìš”ì•½",
  "intent": "job_search | job_change | career_switch | upskilling | study_plan",
  "horizon_months": 1~24,
  "confidence": 0~1,
  "message": "ìƒë‹´ ì‹œì‘ ì•ˆë‚´ 1~2ë¬¸ì¥",
  "next_question": "ì¶”ê°€ íŒŒì•…ì„ ìœ„í•œ êµ¬ì²´ ì§ˆë¬¸ 1ê°œ"
}
- ì—¬ë¶„ ë¬¸ì¥Â·ì´ëª¨ì§€Â·ì„¤ëª… ê¸ˆì§€. ì˜¤ì§ JSON.
- ì£¼ì˜: 'ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤' ê°™ì€ ì„ ì–¸ì  ë¬¸êµ¬ëŠ” ì˜¤ì§ Phase 1ì—ì„œë§Œ 1íšŒ í—ˆìš©.

[Phase 2: 2í„´ ì´í›„]
- ìì—°ì–´ ìƒë‹´ í†¤ìœ¼ë¡œ ë‹µí•´. êµ¬ì¡°: (ê³µê° 1ë¬¸ì¥ â†’ íƒìƒ‰ ì§ˆë¬¸ 1ê°œ â†’ ì‘ê²Œ ì‹¤ì²œí•  ì œì•ˆ 1ë¬¸ì¥)
- ë¶ˆí•„ìš”í•œ ì¥ì‹Â·ì´ëª¨ì§€ ê¸ˆì§€. ê¹”ë”í•˜ê³  êµ¬ì²´ì ìœ¼ë¡œ.
- ì‚¬ìš©ìì˜ í‘œí˜„ì„ ë°˜ì˜í•´ ì¬ì§„ìˆ í•˜ê³ , ìš”ì•½ ë©”ëª¨ë¦¬ë¥¼ ë§¥ë½ìœ¼ë¡œ ì‚¼ì•„ ì¼ê´€ì„± ìœ ì§€.
- ê¸ˆì§€ì–´: "ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ì§„í–‰í•˜ê² ìŠµë‹ˆë‹¤" ë“± ì„ ì–¸ì  ë¬¸êµ¬.

í•­ìƒ í•œêµ­ì–´ë¡œ ë‹µí•´.
""".strip()

# ===== ìœ í‹¸: í˜„ì¬ Phase ëª…ì‹œ =====
def build_phase():
    return "Phase 1 (ì´ˆê¸° í”„ë¡œí•„ JSON)" if state["turn"] == 0 else "Phase 2 (ìƒë‹´í˜• ìì—°ì–´)"

# ===== ìœ í‹¸: ë©”ëª¨ë¦¬ ìš”ì•½ =====
def summarize_memory(history, max_lines=5):
    """
    ëŒ€í™”ë¥¼ 3~5ì¤„ë¡œ ìš”ì•½: ëª©í‘œ/ì œì•½/ì„ í˜¸/ì§„ì²™ ì¤‘ì‹¬, ì„¸ë¶€ ì¡°ì–¸ ì¶•ì•½
    """
    memo_prompt = f"""
    ì•„ë˜ ëŒ€í™”ë¥¼ 3~{max_lines}ì¤„ í•œêµ­ì–´ë¡œ ìš”ì•½í•´.
    ì›ì¹™: 'ëª©í‘œÂ·ì œì•½Â·ì„ í˜¸Â·ì§„ì²™'ë§Œ í¬í•¨, ì„¸ë¶€ ì¡°ì–¸ì€ ì¶•ì•½.
    ê³¼ê±° ìš”ì•½: {state['memory']}

    ëŒ€í™”:
    {history}
    """
    rsp = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0.2,
        messages=[
            {"role":"system","content":"ë„ˆëŠ” ë›°ì–´ë‚œ ìš”ì•½ê°€ë‹¤. ë¶ˆí•„ìš”í•œ ìˆ˜ì‹ ì—†ì´ í•µì‹¬ë§Œ ì“´ë‹¤."},
            {"role":"user","content":memo_prompt}
        ]
    )
    return textwrap.dedent(rsp.choices[0].message.content).strip()

# ===== ìœ í‹¸: Phase 2 ê°€ë“œë ˆì¼(ì„ ì–¸ì  ë¬¸êµ¬ ì œê±°) =====
def clean_phase2_text(s: str) -> str:
    ban_phrases = [
        "ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ì§„í–‰í•˜ê² ìŠµë‹ˆë‹¤",
        "ìƒë‹´ì„ ê³„ì†í•˜ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ë³¸ê²©ì ìœ¼ë¡œ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤"
    ]
    out = s
    for bp in ban_phrases:
        out = out.replace(bp, "").strip()
    while "  " in out:
        out = out.replace("  ", " ")
    return out

# ===== ëª¨ë¸ ì‘ë‹µ =====
def ai_reply(user_text, history_text=""):
    phase = build_phase()
    guide = (
        "[ëŒ€í™” ì§€ì¹¨]\n"
        f"- ì§€ê¸ˆ ë‹¨ê³„: {phase}\n"
        f"- ìš”ì•½ ë©”ëª¨ë¦¬:\n{state['memory']}\n\n"
        "ê·œì¹™:\n"
        "- Phase 1ì´ë©´ ì‹œìŠ¤í…œ ì§€ì‹œëŒ€ë¡œ JSONë§Œ.\n"
        "- Phase 2ë©´ ê³µê°â†’íƒìƒ‰ì§ˆë¬¸(1ê°œ)â†’ì‘ì€ ì‹¤í–‰ ì œì•ˆ(1ê°œ) ìˆœì„œì˜ ìì—°ì–´.\n"
        "- ì§ˆë¬¸ì€ í•œ ë²ˆì— 1ê°œë§Œ, ë‹¨Â·ê°„ê²°Â·êµ¬ì²´.\n"
        "- ê¸ˆì§€ì–´(Phase 2): 'ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤/ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤/ì§„í–‰í•˜ê² ìŠµë‹ˆë‹¤' ë¥˜ì˜ ì„ ì–¸ ë¬¸êµ¬."
    )
    rsp = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0.3 if state["turn"]==0 else 0.4,
        messages=[
            {"role":"system","content":SYSTEM_PROMPT},
            {"role":"assistant","content":guide},
            {"role":"user","content":user_text}
        ]
    )
    out = rsp.choices[0].message.content.strip()

    # Phase 1: JSON ìŠ¤í‚¤ë§ˆ ê°•ì œ
    if state["turn"] == 0:
        try:
            json.loads(out)
        except:
            fix = client.chat.completions.create(
                model="gpt-4o-mini",
                temperature=0.0,
                messages=[
                    {"role":"system","content":"ë‹¤ìŒ í…ìŠ¤íŠ¸ë¥¼ ì§€ì •ëœ JSON ìŠ¤í‚¤ë§ˆë¡œë§Œ ì •ì œí•´. ë‹¤ë¥¸ ë¬¸êµ¬ ê¸ˆì§€."},
                    {"role":"user","content":out}
                ]
            )
            out = fix.choices[0].message.content.strip()

    # ë©”ëª¨ë¦¬ ê°±ì‹ 
    combined = (history_text + f"\nì‚¬ìš©ì: {user_text}\nëª¨ë¸: {out}").strip()
    state["memory"] = summarize_memory(combined)
    state["turn"] += 1

    # Phase 2 í›„ì²˜ë¦¬: ì„ ì–¸ì  ë¬¸êµ¬ ì œê±°
    if state["turn"] > 1:
        out = clean_phase2_text(out)

    return out

# ===== ì½˜ì†” ëŒ€í™” ë£¨í”„ =====
def run_ai_coach():
    print("ì§„ë¡œ ìƒë‹´ ì½”ì¹˜ ì‹œì‘ (exit ì…ë ¥ ì‹œ ì¢…ë£Œ). ì²« í„´ì€ JSON, ì´í›„ ìƒë‹´í˜• ìì—°ì–´.")
    history = ""
    while True:
        user = input("\nì‚¬ìš©ì: ").strip()
        if user.lower() == "exit":
            print("ì¢…ë£Œí•©ë‹ˆë‹¤.")
            break
        bot = ai_reply(user, history_text=history)
        print(f"ì‘ë‹µ:\n{bot}")
        history += f"\nì‚¬ìš©ì: {user}\nëª¨ë¸: {bot}"

# ===== ì‹¤í–‰ =====
if __name__ == "__main__":
    run_ai_coach()
```

```python
# lcel_career_coach.py
from typing import Dict, Any, List
import json, re
from pydantic import BaseModel, Field
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser, JsonOutputParser
from langchain_core.runnables import RunnableLambda, RunnableBranch, RunnablePassthrough

# ====== ì „ì—­ ìƒíƒœ ======
state: Dict[str, Any] = {
    "turn": 0,
    "memory": "",
    "history": []  # [{"role":"user"|"assistant","content":"..."}]
}

# ====== ê³µí†µ LLM ======
# Phase1(JSON) ì•ˆì •ì„± ìœ„í•´ response_format ì‚¬ìš©
llm_json = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.2,
    model_kwargs={"response_format": {"type": "json_object"}}
)
llm_text = ChatOpenAI(model="gpt-4o-mini", temperature=0.5)

# ====== ê¸ˆì§€ì–´ ê°€ë“œë ˆì¼ ======
FORBIDDEN = ["ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤","ìƒë‹´ì„ ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤","ìƒë‹´ì„ ì§„í–‰í•˜ê² ìŠµë‹ˆë‹¤","ìƒë‹´ì„ ê³„ì†í•˜ê² ìŠµë‹ˆë‹¤"]
def clean_forbidden(s: str) -> str:
    out = s
    for w in FORBIDDEN:
        out = out.replace(w, "").strip()
    return re.sub(r"\s{2,}", " ", out)

# ====== Phase1 JSON ìŠ¤í‚¤ë§ˆ ======
class ProfileJSON(BaseModel):
    goal: str = Field(..., description="ì‚¬ìš©ìì˜ 1ì°¨ ì§„ë¡œ ëª©í‘œ í•œ ë¬¸ì¥")
    intent: str = Field(..., description="job_search | job_change | career_switch | upskilling | study_plan")
    horizon_months: int = Field(..., ge=1, le=24)
    confidence: float = Field(..., ge=0.0, le=1.0)
    message: str
    next_question: str

json_parser = JsonOutputParser(pydantic_object=ProfileJSON)
format_instructions = json_parser.get_format_instructions()

# ====== SYSTEM PROMPT ======
SYSTEM_PROMPT = """
ë„ˆëŠ” ë”°ëœ»í•˜ê³  ì „ë¬¸ì ì¸ í•œêµ­ì–´ ì§„ë¡œ ìƒë‹´ ì½”ì¹˜ì•¼.

[Phase 1: ìµœì´ˆ 1í„´]
- ì‚¬ìš©ìì˜ ì§„ë¡œ ëª©í‘œì™€ ìƒí™©ì„ íŒŒì•…í•˜ê¸° ìœ„í•œ "ì´ˆê¸° í”„ë¡œí•„ JSON"ë§Œ ë‹µí•´.
- í‚¤: goal, intent, horizon_months, confidence, message, next_question
- ì—¬ë¶„ ë¬¸ì¥Â·ì´ëª¨ì§€Â·ì„¤ëª… ê¸ˆì§€. ì˜¤ì§ JSON.

[Phase 2: 2í„´ ì´í›„]
- ìì—°ì–´ ìƒë‹´ í†¤: ê³µê° 1ë¬¸ì¥ â†’ íƒìƒ‰ ì§ˆë¬¸ 1ê°œ â†’ ì‘ê²Œ ì‹¤ì²œí•  ì œì•ˆ 1ë¬¸ì¥.
- ì¥ì‹ ê¸ˆì§€, êµ¬ì²´ì Â·ê°„ê²°.
- ì´ì „ ëŒ€í™” ë§¥ë½ì„ ë°˜ì˜í•´ ì¼ê´€ì„± ìœ ì§€.
- ê¸ˆì§€ì–´: "ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤" ë“± ì„ ì–¸ë¬¸.
í•­ìƒ í•œêµ­ì–´ë¡œ.
""".strip()

def build_guide() -> str:
    phase = "Phase 1 (ì´ˆê¸° í”„ë¡œí•„ JSON)" if state["turn"] == 0 else "Phase 2 (ìƒë‹´í˜• ìì—°ì–´)"
    return (
        f"[í˜„ì¬ ë‹¨ê³„: {phase}]\n"
        f"[í„´: {state['turn'] + 1}]\n"
        f"[ë©”ëª¨ë¦¬: {state['memory']}]\n\n"
        "- Phase 1(í„´=1): JSONë§Œ ì¶œë ¥\n"
        "- Phase 2(í„´â‰¥2): ê³µê°â†’ì§ˆë¬¸ 1ê°œâ†’ì œì•ˆ\n"
        "- ì´ì „ ëŒ€í™”ë¥¼ ì°¸ê³ í•´ ë§¥ë½ ìœ ì§€"
    )

def to_chat_messages(history: List[Dict[str, str]]) -> List[Dict[str, str]]:
    return history

# ====== í”„ë¡¬í”„íŠ¸ë“¤ ======
phase1_prompt = ChatPromptTemplate.from_messages([
    ("system", SYSTEM_PROMPT),
    ("system", "{guide}"),
    MessagesPlaceholder("history"),
    ("human", "{user_input}"),
    ("system", "ì•„ë˜ ìŠ¤í‚¤ë§ˆë¥¼ ì •í™•íˆ ë”°ë¥´ëŠ” JSONë§Œ ì¶œë ¥:\n{format_instructions}")
])

phase2_prompt = ChatPromptTemplate.from_messages([
    ("system", SYSTEM_PROMPT),
    ("system", "{guide}"),
    MessagesPlaceholder("history"),
    ("human", "{user_input}"),
    ("system", "ì¶œë ¥ì€ ì •í™•íˆ 3ë¬¸ì¥: ê³µê° 1ë¬¸ì¥ â†’ íƒìƒ‰ ì§ˆë¬¸ 1ê°œ â†’ ì‘ê²Œ ì‹¤ì²œí•  ì œì•ˆ 1ë¬¸ì¥. ê¸ˆì§€ì–´ ì‚¬ìš© ê¸ˆì§€.")
])

# ====== ìš”ì•½ ì²´ì¸(5í„´ë§ˆë‹¤) ======
summary_prompt = ChatPromptTemplate.from_template(
    "ì•„ë˜ ëŒ€í™”ë¥¼ 3~5ì¤„ë¡œ ìš”ì•½í•´. 'ëª©í‘œÂ·ì œì•½Â·ì„ í˜¸Â·ì§„ì²™'ë§Œ í¬í•¨, ì„¸ë¶€ ì¡°ì–¸ì€ ì¶•ì•½.\n\n{history_text}"
)
summary_chain = (summary_prompt | llm_text | StrOutputParser())

def summarize_and_trim():
    history_text = "\n".join([f"{m['role']}: {m['content']}" for m in state["history"]])
    memo = summary_chain.invoke({"history_text": history_text}).strip()
    state["memory"] = memo
    if len(state["history"]) > 16:
        state["history"] = state["history"][-8:]

# ====== Phaseë³„ ì‹¤í–‰ ì²´ì¸ ======
phase1_chain = (
    RunnablePassthrough.assign(
        guide=lambda x: build_guide(),
        history=lambda x: to_chat_messages(state["history"]),
        format_instructions=lambda x: format_instructions
    )
    | phase1_prompt
    | llm_json
    | StrOutputParser()                               # ë¬¸ìì—´ë¡œ ìˆ˜ì‹ 
    | RunnableLambda(lambda s: json.dumps(json.loads(s), ensure_ascii=False))  # JSON ì •í•© ë³´ì¦
)

phase2_chain = (
    RunnablePassthrough.assign(
        guide=lambda x: build_guide(),
        history=lambda x: to_chat_messages(state["history"]),
    )
    | phase2_prompt
    | llm_text
    | StrOutputParser()
    | RunnableLambda(clean_forbidden)
)

# ====== ë¼ìš°í„°(Phase ë¶„ê¸°) ======
router = RunnableBranch(
    (lambda x: state["turn"] == 0, phase1_chain),
    phase2_chain
)

# ====== í•œ í„´ ì‹¤í–‰ ======
def run_turn(user_input: str) -> str:
    out = router.invoke({"user_input": user_input})
    # íˆìŠ¤í† ë¦¬ ì ì¬
    state["history"].append({"role": "user", "content": user_input})
    state["history"].append({"role": "assistant", "content": out})
    # í„´ ì¦ê°€
    state["turn"] += 1
    # 5í„´ë§ˆë‹¤ ë©”ëª¨ë¦¬ ìš”ì•½ + íˆìŠ¤í† ë¦¬ íŠ¸ë¦¬ë°
    if state["turn"] > 0 and state["turn"] % 5 == 0:
        summarize_and_trim()
    return out

# ====== ì½˜ì†” ë£¨í”„ ======
def main():
    print("ì§„ë¡œ ìƒë‹´ ì½”ì¹˜ (ì¢…ë£Œ: exit)\nì²« í„´: JSON í”„ë¡œí•„ / ì´í›„: ìì—°ì–´ ìƒë‹´")
    while True:
        u = input("\nì‚¬ìš©ì: ").strip()
        if u.lower() == "exit":
            print("ì¢…ë£Œí•©ë‹ˆë‹¤."); break
        a = run_turn(u)
        print("\nì‘ë‹µ:\n", a)
        print(f"\n[ë””ë²„ê·¸] turn={state['turn']}, history={len(state['history'])}, mem_len={len(state['memory'])}")

if __name__ == "__main__":
    main()
```

```python
# lcel_career_coach.py
from typing import Dict, Any, List
import json, re
from pydantic import BaseModel, Field
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser, JsonOutputParser
from langchain_core.runnables import RunnableLambda, RunnableBranch, RunnablePassthrough

# ====== ì „ì—­ ìƒíƒœ ======
state: Dict[str, Any] = {
    "turn": 0,
    "memory": "",
    "history": []  # [{"role":"user"|"assistant","content":"..."}]
}

# ====== ê³µí†µ LLM ======
# Phase1(JSON) ì•ˆì •ì„± ìœ„í•´ response_format ì‚¬ìš©
llm_json = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.2,
    model_kwargs={"response_format": {"type": "json_object"}}
)
llm_text = ChatOpenAI(model="gpt-4o-mini", temperature=0.5)

# ====== ê¸ˆì§€ì–´ ê°€ë“œë ˆì¼ ======
FORBIDDEN = ["ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤","ìƒë‹´ì„ ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤","ìƒë‹´ì„ ì§„í–‰í•˜ê² ìŠµë‹ˆë‹¤","ìƒë‹´ì„ ê³„ì†í•˜ê² ìŠµë‹ˆë‹¤"]
def clean_forbidden(s: str) -> str:
    out = s
    for w in FORBIDDEN:
        out = out.replace(w, "").strip()
    return re.sub(r"\s{2,}", " ", out)

# ====== Phase1 JSON ìŠ¤í‚¤ë§ˆ ======
class ProfileJSON(BaseModel):
    goal: str = Field(..., description="ì‚¬ìš©ìì˜ 1ì°¨ ì§„ë¡œ ëª©í‘œ í•œ ë¬¸ì¥")
    intent: str = Field(..., description="job_search | job_change | career_switch | upskilling | study_plan")
    horizon_months: int = Field(..., ge=1, le=24)
    confidence: float = Field(..., ge=0.0, le=1.0)
    message: str
    next_question: str

json_parser = JsonOutputParser(pydantic_object=ProfileJSON)
format_instructions = json_parser.get_format_instructions()

# ====== SYSTEM PROMPT ======
SYSTEM_PROMPT = """
ë„ˆëŠ” ë”°ëœ»í•˜ê³  ì „ë¬¸ì ì¸ í•œêµ­ì–´ ì§„ë¡œ ìƒë‹´ ì½”ì¹˜ì•¼.

[Phase 1: ìµœì´ˆ 1í„´]
- ì‚¬ìš©ìì˜ ì§„ë¡œ ëª©í‘œì™€ ìƒí™©ì„ íŒŒì•…í•˜ê¸° ìœ„í•œ "ì´ˆê¸° í”„ë¡œí•„ JSON"ë§Œ ë‹µí•´.
- í‚¤: goal, intent, horizon_months, confidence, message, next_question
- ì—¬ë¶„ ë¬¸ì¥Â·ì´ëª¨ì§€Â·ì„¤ëª… ê¸ˆì§€. ì˜¤ì§ JSON.

[Phase 2: 2í„´ ì´í›„]
- ìì—°ì–´ ìƒë‹´ í†¤: ê³µê° 1ë¬¸ì¥ â†’ íƒìƒ‰ ì§ˆë¬¸ 1ê°œ â†’ ì‘ê²Œ ì‹¤ì²œí•  ì œì•ˆ 1ë¬¸ì¥.
- ì¥ì‹ ê¸ˆì§€, êµ¬ì²´ì Â·ê°„ê²°.
- ì´ì „ ëŒ€í™” ë§¥ë½ì„ ë°˜ì˜í•´ ì¼ê´€ì„± ìœ ì§€.
- ê¸ˆì§€ì–´: "ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤" ë“± ì„ ì–¸ë¬¸.
í•­ìƒ í•œêµ­ì–´ë¡œ.
""".strip()

def build_guide() -> str:
    phase = "Phase 1 (ì´ˆê¸° í”„ë¡œí•„ JSON)" if state["turn"] == 0 else "Phase 2 (ìƒë‹´í˜• ìì—°ì–´)"
    return (
        f"[í˜„ì¬ ë‹¨ê³„: {phase}]\n"
        f"[í„´: {state['turn'] + 1}]\n"
        f"[ë©”ëª¨ë¦¬: {state['memory']}]\n\n"
        "- Phase 1(í„´=1): JSONë§Œ ì¶œë ¥\n"
        "- Phase 2(í„´â‰¥2): ê³µê°â†’ì§ˆë¬¸ 1ê°œâ†’ì œì•ˆ\n"
        "- ì´ì „ ëŒ€í™”ë¥¼ ì°¸ê³ í•´ ë§¥ë½ ìœ ì§€"
    )

def to_chat_messages(history: List[Dict[str, str]]) -> List[Dict[str, str]]:
    return history

# ====== í”„ë¡¬í”„íŠ¸ë“¤ ======
phase1_prompt = ChatPromptTemplate.from_messages([
    ("system", SYSTEM_PROMPT),
    ("system", "{guide}"),
    MessagesPlaceholder("history"),
    ("human", "{user_input}"),
    ("system", "ì•„ë˜ ìŠ¤í‚¤ë§ˆë¥¼ ì •í™•íˆ ë”°ë¥´ëŠ” JSONë§Œ ì¶œë ¥:\n{format_instructions}")
])

phase2_prompt = ChatPromptTemplate.from_messages([
    ("system", SYSTEM_PROMPT),
    ("system", "{guide}"),
    MessagesPlaceholder("history"),
    ("human", "{user_input}"),
    ("system", "ì¶œë ¥ì€ ì •í™•íˆ 3ë¬¸ì¥: ê³µê° 1ë¬¸ì¥ â†’ íƒìƒ‰ ì§ˆë¬¸ 1ê°œ â†’ ì‘ê²Œ ì‹¤ì²œí•  ì œì•ˆ 1ë¬¸ì¥. ê¸ˆì§€ì–´ ì‚¬ìš© ê¸ˆì§€.")
])

# ====== ìš”ì•½ ì²´ì¸(5í„´ë§ˆë‹¤) ======
summary_prompt = ChatPromptTemplate.from_template(
    "ì•„ë˜ ëŒ€í™”ë¥¼ 3~5ì¤„ë¡œ ìš”ì•½í•´. 'ëª©í‘œÂ·ì œì•½Â·ì„ í˜¸Â·ì§„ì²™'ë§Œ í¬í•¨, ì„¸ë¶€ ì¡°ì–¸ì€ ì¶•ì•½.\n\n{history_text}"
)
summary_chain = (summary_prompt | llm_text | StrOutputParser())

def summarize_and_trim():
    history_text = "\n".join([f"{m['role']}: {m['content']}" for m in state["history"]])
    memo = summary_chain.invoke({"history_text": history_text}).strip()
    state["memory"] = memo
    if len(state["history"]) > 16:
        state["history"] = state["history"][-8:]

# ====== Phaseë³„ ì‹¤í–‰ ì²´ì¸ ======
phase1_chain = (
    RunnablePassthrough.assign(
        guide=lambda x: build_guide(),
        history=lambda x: to_chat_messages(state["history"]),
        format_instructions=lambda x: format_instructions
    )
    | phase1_prompt
    | llm_json
    | StrOutputParser()                               # ë¬¸ìì—´ë¡œ ìˆ˜ì‹ 
    | RunnableLambda(lambda s: json.dumps(json.loads(s), ensure_ascii=False))  # JSON ì •í•© ë³´ì¦
)

phase2_chain = (
    RunnablePassthrough.assign(
        guide=lambda x: build_guide(),
        history=lambda x: to_chat_messages(state["history"]),
    )
    | phase2_prompt
    | llm_text
    | StrOutputParser()
    | RunnableLambda(clean_forbidden)
)

# ====== ë¼ìš°í„°(Phase ë¶„ê¸°) ======
router = RunnableBranch(
    (lambda x: state["turn"] == 0, phase1_chain),
    phase2_chain
)

# ====== í•œ í„´ ì‹¤í–‰ ======
def run_turn(user_input: str) -> str:
    out = router.invoke({"user_input": user_input})
    # íˆìŠ¤í† ë¦¬ ì ì¬
    state["history"].append({"role": "user", "content": user_input})
    state["history"].append({"role": "assistant", "content": out})
    # í„´ ì¦ê°€
    state["turn"] += 1
    # 5í„´ë§ˆë‹¤ ë©”ëª¨ë¦¬ ìš”ì•½ + íˆìŠ¤í† ë¦¬ íŠ¸ë¦¬ë°
    if state["turn"] > 0 and state["turn"] % 5 == 0:
        summarize_and_trim()
    return out

# ====== ì½˜ì†” ë£¨í”„ ======
def main():
    print("ì§„ë¡œ ìƒë‹´ ì½”ì¹˜ (ì¢…ë£Œ: exit)\nì²« í„´: JSON í”„ë¡œí•„ / ì´í›„: ìì—°ì–´ ìƒë‹´")
    while True:
        u = input("\nì‚¬ìš©ì: ").strip()
        if u.lower() == "exit":
            print("ì¢…ë£Œí•©ë‹ˆë‹¤."); break
        a = run_turn(u)
        print("\nì‘ë‹µ:\n", a)
        print(f"\n[ë””ë²„ê·¸] turn={state['turn']}, history={len(state['history'])}, mem_len={len(state['memory'])}")

if __name__ == "__main__":
    main()
```

```python
import json, re
from typing import Dict, Any, List
from pydantic import BaseModel, Field
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser, JsonOutputParser
from langchain_core.runnables import RunnableLambda, RunnableBranch, RunnablePassthrough

# ===== ì „ì—­ ìƒíƒœ =====
state: Dict[str, Any] = {"turn": 0, "memory": "", "history": []}
```

```python
# ===== ëª¨ë¸ ì„¤ì • =====
llm_json = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.2,
    model_kwargs={"response_format": {"type": "json_object"}}
)
llm_text = ChatOpenAI(model="gpt-4o-mini", temperature=0.5)

# ===== ê¸ˆì§€ì–´ í•„í„° =====
FORBIDDEN = [
    "ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤",
    "ìƒë‹´ì„ ì§„í–‰í•˜ê² ìŠµë‹ˆë‹¤", "ìƒë‹´ì„ ê³„ì†í•˜ê² ìŠµë‹ˆë‹¤"
]
def clean_forbidden(s: str) -> str:
    for w in FORBIDDEN:
        s = s.replace(w, "")
    return re.sub(r"\s{2,}", " ", s).strip()
```

```python
# ===== Phase 1 JSON êµ¬ì¡° =====
class ProfileJSON(BaseModel):
    goal: str
    intent: str
    horizon_months: int
    confidence: float
    message: str
    next_question: str

json_parser = JsonOutputParser(pydantic_object=ProfileJSON)
format_instructions = json_parser.get_format_instructions()

SYSTEM_PROMPT = """
ë„ˆëŠ” ë”°ëœ»í•˜ê³  ì „ë¬¸ì ì¸ ì§„ë¡œ ìƒë‹´ ì½”ì¹˜ì•¼.
[Phase 1: ì²« ëŒ€í™”]
- ì‚¬ìš©ìì˜ ìƒí™©ì„ ë¶„ì„í•˜ê³  JSONë§Œ ì¶œë ¥í•´.
- í‚¤: goal, intent, horizon_months, confidence, message, next_question
- ì„¤ëª…ë¬¸, ì´ëª¨ì§€, ì¥ì‹ ê¸ˆì§€.
""".strip()

def build_guide():
    phase = "Phase 1 (í”„ë¡œí•„ ìˆ˜ì§‘)" if state["turn"] == 0 else "Phase 2 (ìƒë‹´)"
    return f"[í˜„ì¬ ë‹¨ê³„: {phase}] [í„´: {state['turn']+1}] [ìš”ì•½: {state['memory']}]"

phase1_prompt = ChatPromptTemplate.from_messages([
    ("system", SYSTEM_PROMPT),
    ("system", "{guide}"),
    MessagesPlaceholder("history"),
    ("human", "{user_input}"),
    ("system", "ì¶œë ¥ì€ ì•„ë˜ í˜•ì‹ì„ ë°˜ë“œì‹œ ë”°ë¥¼ ê²ƒ:\n{format_instructions}")
])

phase1_chain = (
    RunnablePassthrough.assign(
        guide=lambda x: build_guide(),
        history=lambda x: state["history"],
        format_instructions=lambda x: format_instructions
    )
    | phase1_prompt
    | llm_json
    | StrOutputParser()
    | RunnableLambda(lambda s: json.dumps(json.loads(s), ensure_ascii=False))
)
```

```python
SYS_PHASE2 = """
ë„ˆëŠ” ë”°ëœ»í•˜ê³  ì „ë¬¸ì ì¸ ì§„ë¡œ ìƒë‹´ ì½”ì¹˜ì•¼.
[Phase 2: 2í„´ ì´í›„]
- ì¶œë ¥ì€ ì •í™•íˆ 3ë¬¸ì¥ (ê³µê° â†’ íƒìƒ‰ ì§ˆë¬¸ â†’ ì‘ê²Œ ì‹¤ì²œí•  ì œì•ˆ)
- ê¸ˆì§€ì–´: 'ìƒë‹´ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤', 'ìƒë‹´ì„ ì´ì–´ê°€ê² ìŠµë‹ˆë‹¤', 'ìƒë‹´ì„ ì§„í–‰í•˜ê² ìŠµë‹ˆë‹¤', 'ìƒë‹´ì„ ê³„ì†í•˜ê² ìŠµë‹ˆë‹¤'
""".strip()

phase2_prompt = ChatPromptTemplate.from_messages([
    ("system", SYS_PHASE2),
    ("system", "{guide}"),
    MessagesPlaceholder("history"),
    ("human", "{user_input}")
])

phase2_chain = (
    RunnablePassthrough.assign(
        guide=lambda x: build_guide(),
        history=lambda x: state["history"]
    )
    | phase2_prompt
    | llm_text
    | StrOutputParser()
    | RunnableLambda(clean_forbidden)
)

# ===== ìš”ì•½ ì²´ì¸ =====
summary_prompt = ChatPromptTemplate.from_template(
    "ë‹¤ìŒ ëŒ€í™”ë¥¼ ìš”ì•½í•´ì¤˜. 'ëª©í‘œ, ì œì•½, ì„ í˜¸, ì§„ì²™' ì¤‘ì‹¬ìœ¼ë¡œ 3~5ì¤„:\n\n{history_text}"
)
summary_chain = summary_prompt | llm_text | StrOutputParser()

def summarize_and_trim():
    """5í„´ë§ˆë‹¤ Memory ì—…ë°ì´íŠ¸"""
    history_text = "\n".join([f"{m['role']}: {m['content']}" for m in state["history"]])
    state["memory"] = summary_chain.invoke({"history_text": history_text}).strip()
    if len(state["history"]) > 16:
        state["history"] = state["history"][-8:]
```

```python
# Phase ì „í™˜ ë¼ìš°í„°
router = RunnableBranch(
    (lambda x: state["turn"] == 0, phase1_chain),
    phase2_chain
)

def run_turn(user_input: str) -> str:
    """Phaseë³„ ì‹¤í–‰ + ì˜ˆì™¸ ë³µêµ¬"""
    try:
        out = router.invoke({"user_input": user_input})
        if state["turn"] == 0:
            try:
                json.loads(out)
            except Exception:
                out = json.dumps({"goal": "ë¯¸ì •", "intent": "study_plan", "horizon_months": 3,
                                  "confidence": 0.5, "message": "í˜•ì‹ ì˜¤ë¥˜ë¡œ ê¸°ë³¸ê°’ ì ìš©",
                                  "next_question": "í˜„ì¬ ê´€ì‹¬ ë¶„ì•¼ëŠ” ë¬´ì—‡ì¸ê°€ìš”?"}, ensure_ascii=False)
        state["history"].append({"role": "user", "content": user_input})
        state["history"].append({"role": "assistant", "content": out})
        state["turn"] += 1
        if state["turn"] % 5 == 0:
            summarize_and_trim()
        return out
    except Exception as e:
        return f"âš ï¸ ì˜¤ë¥˜ ë°œìƒ: {e}"
```
