```python
# 리스트 메서드 1 - append - 맨 뒤에 값을 추가하는 기능

nums = [1,2,3]
```

```python
print(nums)
```

```python
# 4라는 숫자를 맨 뒤에 추가
nums.append(4)
print(nums)
```

```python
# 2) insert - 원하는 위치에 값을 추가 insert(위치, 값)
# [1,2,3,4,4]에서 맨 뒤의 4를 5로 삽입

# 4번째 위치의 값을 5로 삽입해줘
nums.insert(4,5)
print(nums)
```

```python
# 3) remove(값) - 해당 값을 찾아서 삭제
# remove는 값을 순차적으로 1개씩 제거를 한다.
nums.remove(4)
print(nums)
```

```python
# 4) pop 은 마지막 요소를 제거
# 1) 제거된 값도 가지고 있다.
# 2) 남은 리스트도 출력할 수 있다.
deleted = nums.pop()
print(deleted)
print(nums)
```

```python
# 5) 정렬 - sort() - 오름차순 (작은 값 -> 큰값) / 내림차순(큰값 -> 작은값)
# 기본적으로는 오름차순 / 내림차순(reverse = True)

nums.sort(reverse=True) # 내림차순
print(nums)
```

```python
# 6) count()와 index() - 탐색
# count - 등장 횟수 ex) 5가 몇개 인지? 3이 몇개인지?
# index - 데이터의 위치

print(nums.count(4)) # 데이터가 몇개 있는지 파악! 없으면 0
print(nums.index(3)) # 데이터의 위치를 파악하는데 용이
```

```python
# 리스트의 갯수를 셀때 len을 사용하면 리스트의 총 길이를 알수 잇다.(데이터의 갯수를 알 수 있다.)
print(len(nums))
```

```python
# 1. 반복 구조에 대한 뼈대
# 반복을 하는 이유가 있다. -> 단어를 지속적으로 입력을 받는다. 데이터가 다 처리될때까지 종료하지 않는다.
# 프로그램에 대한 시작 / 유지

while True: # 조건이 만족할때는 계속 실행해라!
    word = input("단어를 입력하세요 (종료: exit를 입력하세요)")
```

```python
# 2. while같은 경우 특히 종료에 대한 조건을 추가하는 것이 중요
# 왜냐면 while은 특수한 상황이 아니면 종료를 하지 않는 반복문이기 때문
# 프로그램에 대한 종료

while True: # 조건이 만족할때는 계속 실행해라!
    word = input("단어를 입력하세요 (종료: exit를 입력하세요)")
    if word == 'exit':
        break # exit를 입력 받으면 종료해라!
```

```python
# 4. 단어를 추가하기 위한 리스트가 필요
words = []

# 3. 빈 문자열에 대한 예외 처리 (잘못 입력된 문자열)
while True: # 조건이 만족할때는 계속 실행해라!
    word = input("단어를 입력하세요 (종료: exit를 입력하세요)")

    if word == 'exit':
        break # exit를 입력 받으면 종료해라!

    elif word == '':
        print("공백은 입력할 수 없습니다.")
# 4. 위의 조건에 부합하면 -> 그대로 단어를 저장하는 (리스트에) 구문이 필요
# append를 사용해서 단어를 추가
    else:
        words.append(word) # 내부에서 작동
        print(f"{word}가 추가되었습니다.") # 고객에게 표현

print("입력된 단어 목록은 무엇인가요?")
for w in words:
    print(w)
```
