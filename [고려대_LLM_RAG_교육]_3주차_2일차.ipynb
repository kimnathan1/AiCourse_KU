````python
# ChatGPT ëª¨ë¸ êµ¬í˜„
def ask(query):
    # ì‚¬ìš©ì ì…ë ¥ -> ë©”ì„¸ì§€ ëª©ë¡ ì´ˆê¸°í™” -> ì´ í˜•íƒœë¡œ ëŒ€í™”ê°€ ì‹œì‘ë˜ê³  íˆìŠ¤í† ë¦¬ê°€ ì €ì¥
    # ëŒ€í™”ì˜ ì²«ë²ˆì§¸ í„´ì„ êµ¬ì„±! -> íˆìŠ¤í† ë¦¬ ì‹œì‘!
    # ChatGPT API ë¦¬ìŠ¤íŠ¸ ë°©ì‹ìœ¼ë¡œ ëŒ€í™”ë¥¼ ì£¼ê³ ë°›ëŠ” ê²ƒì— ìµìˆ™
    # ëˆ„ê°€ ë§í–ˆê³  ë¬´ì—‡ì„ ë§í–ˆëŠ”ì§€ë¥¼ ì§€ì†ì ìœ¼ë¡œ ì €ì¥
    # query -> ì˜¤ëŠ˜ 2025ë…„ë„ 10ì›” 28ì¼ì˜ ì›-ë‹¬ëŸ¬ í™˜ìœ¨ì„ ì•Œë ¤ì¤˜
    messages = [{"role" : "user" , "content" : query}]

    while True:
        resp = client.chat.completions.create(
            model = 'gpt-4o-mini',
            messages=messages,
            tools = TOOLS, # ìœ„ì—ì„œ ì •ì˜í•œ TOOLSë¼ëŠ” JSON ëª…ì„¸ -> ì´ëŸ°ë„êµ¬ê°€ ìˆìœ¼ë‹ˆ ì‚¬ìš©í•´ë¼!
            tool_choice='auto', # ëª¨ë¸ì´ í•„ìš”í•œ ê²½ìš° ìë™ìœ¼ë¡œ í•¨ìˆ˜ë¥¼ í˜¸ì¶œ!
            temperature = 0.2 # í™˜ìœ¨ ì •ë³´ëŠ” ê³ ì •! -> ì˜¤ëŠ˜ í™˜ìœ¨ì€ 1320ì›
        )
        msg = resp.choices[0].message
        # tool_callsì— ëŒ€í•œ (í•¨ìˆ˜ì— ëŒ€í•œ) í˜¸ì¶œì´ ì—†ìœ¼ë©´ ë°”ë¡œ ì¢…ë£Œ í•´ë¼
        if not msg.tool_calls:
            return msg.content

        # ì–´ì‹œìŠ¤í„´íŠ¸ ë©”ì„¸ì§€(í˜¸ì¶œì— ëŒ€í•œ ì˜ë„) ì¶”ê°€
        # ì´ì œ í•¨ìˆ˜ë¥¼ ì‹¤ì œë¡œ ì‹¤í–‰í•´ì•¼ë˜ìš” ë¼ê³  ìš”ì²­ì„ ë³´ëƒˆì„ë•Œ ëŒ€í™” ê¸°ë¡ì„ ì €ì¥!
        # í™˜ìœ¨ì— ëŒ€í•œ ì •ë³´ í˜¸ì¶œ -> ëŒ€í™”ì˜ ë§¥ë½ì´ ìœ ì§€ê°€ ëœë‹¤.
        messages.append({
            "role" : "assistant",
            "content" : msg.content or "",
            "tool_calls" : msg.tool_calls
        })

        # ì‹¤ì œ í•¨ìˆ˜ê°€ ì‹¤í–‰ë˜ëŠ” ë¶€ë¶„ ë° ê²°ê³¼ ì²¨ë¶€
        # ë„êµ¬ ì‹¤í–‰ -> ê²°ê³¼ë¥¼ tool ë©”ì„¸ì§€ë¡œ ì²¨ë¶€(ì¤‘ìš” : tool_call_id)

        for tc in msg.tool_calls:
            args = json.loads(tc.function.arguments or "{}")
            result = get_exchange_rate(**args)
            messages.append({
                "role" : "tool",
                "tool_call_id" : tc.id, # í•„ìˆ˜
                "name" : tc.function.name, # ê°€ë…ì„±
                "content" : result # JSON ë¬¸ìì—´
            })
````
